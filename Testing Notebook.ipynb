## Testing

```python
model = keras.models.load_model('./Model/Fitted_Model.h5')
```

```python
def apply_image_trans(img):
    img_size = 224
    final_image = cv2.resize(img, (img_size, img_size))
    final_image = np.expand_dims(final_image, axis = 0)
    final_image = final_image/255.0
    
    return final_image

# Function for interpreting the output
def convert_sigmoid_output(sigmoid):
    if(sigmoid < 0.50):
        print('\nLouis Has Been Spotted!!!!\n')
    elif(sigmoid > 0.50):
        print('\nLouis Is Not Present!\n')
    else:
        print('\nNot Sure!\n')
```

#### Quick note: the trained classifier actually hasn't seen these images before, so there is no data leakage happening
